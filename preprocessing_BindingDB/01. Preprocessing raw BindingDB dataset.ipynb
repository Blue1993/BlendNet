### 03. Split to IC50 and Ki datasets

```python
IC50_df = BindingDB_interactions[BindingDB_interactions["IC50 (nM)"].isnull() == False].reset_index(drop = True)
IC50_df = IC50_df[(IC50_df["IC50 (nM)"].str.contains(">") == False) & (IC50_df["IC50 (nM)"].str.contains("<") == False)].reset_index(drop = True)
IC50_df
```

```python
Ki_df = BindingDB_interactions[BindingDB_interactions["Ki (nM)"].isnull() == False].reset_index(drop = True)
Ki_df = Ki_df[(Ki_df["Ki (nM)"].str.contains(">") == False)  & (Ki_df["Ki (nM)"].str.contains("<") == False)].reset_index(drop = True)
Ki_df
```

```python
def extract_info(df, type_):
    convert_dict = {"Reactant_set_id":[],"UniProt_ids":[], "CID":[], "Labels":[], "SMILES":[], "Seqs":[]}
    type_index_dict = {"Ki":8, "IC50":9, "Kd":10, "EC50":11}
    
    for line in df.values:
        type_index = type_index_dict[type_]
        Reactant_id, Uniprot_id, CID, Seqs, smiles, aff = line[0], line[42], int(line[29]), line[38], line[1], line[type_index]
        
        if "[" in Uniprot_id:
            tmp = Uniprot_id.split(",")
            if "]" not in tmp[0]:
                Uniprot_id = tmp[0] + "]"
            elif "]" in tmp[0]:
                Uniprot_id = tmp[0]
        else:
            tmp = Uniprot_id.split(" ")
            if len(tmp) != 1:
                continue
                
        convert_dict["Reactant_set_id"].append(Reactant_id)
        convert_dict["UniProt_ids"].append(Uniprot_id)
        convert_dict["CID"].append(CID)
        convert_dict["Labels"].append(float(aff.strip()))
        convert_dict["SMILES"].append(smiles)
        convert_dict["Seqs"].append(Seqs.upper())

    return pd.DataFrame(convert_dict)
```

```python
obsolete_dict = {"Q66HS9":"Q63344", "Q7KAN4":"O61142", "Q8MU52":"Q8ILQ7"}
```

```python
def get_seq(df, retirve_dict, obsolete_dict):
    results = list()
    for line in df.values:
        id_ = line[1]
        
        if "[" in id_:
            start, end = int(id_.split("[")[1].strip("]").split("-")[0]), int(id_.split("[")[1].strip("]").split("-")[1])
            id_ = id_.split("[")[0]
            results.append(retirve_dict[id_][start-1:end-1])
            
        elif id_ in obsolete_dict:
            id_ = obsolete_dict[id_]
            results.append(retirve_dict[id_])
        
        else:
            results.append(retirve_dict[id_])
    return results
```

### 05. Merging duplicate interactions

```python
# nM unit -> 1e-9
def log_transform(val):
    return - np.log10(val * 1e-9)

# 10000000. = 2, 0.001 = 12
def check_afordable_range(input):
    for idx, i in enumerate(input):
        if i > 1e+7: 
            input[idx] = 1e+7
        elif i < 0.001:
            input[idx] = 1e-3
    
    return input

def merge(df):
    df_groups = df.groupby(["UniProt_ids", "CID"])
    print(f"\tUnique interactions:  groupby: {len(df_groups.groups)}")
    
    convert_dict = {"Seqs":[], "CID":[], "Labels":[], "Reactant_set_id":[], "SMILES":[], "UniProt_ids":[]}
    
    for i in df_groups.groups:
        flag = 0
        uniprot_id, CID = i[0], i[1]
        index = df_groups.groups[i]
        
        reactant_id = df.iloc[index, 0].values
        Seqs = df.iloc[index, 5].values[0]
        aff = df.iloc[index, 3].values
        smiles = df.iloc[index, 4].values

        aff = check_afordable_range(aff)
        
        if np.max(aff) <= 100:
            aff = np.median(aff)
            aff = log_transform(aff)
            flag = 1
            
        elif np.min(aff) * 100 >= np.max(aff): 
            aff = np.median(aff)
            aff = log_transform(aff)
            flag = 1
        
        if flag == 1:
            flag = 0
            convert_dict["UniProt_ids"].append(uniprot_id)
            convert_dict["CID"].append(CID)
            convert_dict["Labels"].append(aff)
            
            reactant_id = list(map(str, reactant_id))
            convert_dict["Reactant_set_id"].append(",".join(reactant_id))
            convert_dict["SMILES"].append(",".join(smiles))
            convert_dict["Seqs"].append(Seqs)

    return pd.DataFrame(convert_dict)
```
